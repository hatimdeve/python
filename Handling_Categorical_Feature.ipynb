**Ordinal Number Encoding**

```python
import datetime
```

```python
today_date=datetime.datetime.today()
today_date
```

```python
#### List Comprehension
days=[today_date-datetime.timedelta(x) for x in range(0,15)]
days
```

```python
days[0].day_name()
```

```python
import pandas as pd
data=pd.DataFrame(days)
data.columns=["Day"]
data.head()
```

```python
data.info()
```

```python
data['weekday_ordinal']=data['weekday'].map(dictionary)
data.head()
```

**Count Or Frequency Encoding**

```python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
df.head()              
```

```python
columns=[1,3,5,6,7,8,9,13]
df=df[columns]
df.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
df.head()
```

```python
for feature in df.columns[:]:
    print(feature,":",len(df[feature].unique()),'labels')
```

```python
country_map=df['Country'].value_counts().to_dict()
```

```python
df['Country']=df['Country'].map(country_map)
df.head(20)
```

**Advantages**
* Easy To Use
* Not increasing feature space

**Disadvantages**

* It will provide same weight if the frequencies are same

**Target Guided Ordinal Encoding**

* Ordering the labels according to the target
* Replace the labels by the joint probability of being 1 or 0

```python
from google.colab import files
import io
uploaded =files.upload()
```

```python
import pandas as pd
df=pd.read_csv(io.BytesIO(uploaded['titanic.csv']),usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

**Mean Encoding**

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```

**Probability Ratio Encoding**

* Probability of Survived based on Cabin--- Categorical Feature
* Probability of Not Survived---1-pr(Survived)
* pr(Survived)/pr(Not Survived)
* Dictonary to map cabin with probability
* replace with the categorical feature

```python
df=pd.read_csv(io.BytesIO(uploaded['titanic.csv']),usecols=['Cabin','Survived'])
df.head()
```

```python
### Replacing
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived']
prob_df.head()
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
prob_df.head()
```

```python
probability_encoded=prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded']=df['Cabin'].map(probability_encoded)
df.head()
```
