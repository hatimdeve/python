```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'x':np.random.normal(0,2,1000),
                   'y':np.random.normal(3,5,1000),
                   'z':np.random.normal(-4,10,1000)})
```

```python
sns.set()
plt.figure(figsize=(4,3))
for col in df.columns:
  plt.hist(df[col],density=True,alpha=0.5,bins=30)
  sns.kdeplot(df[col],shade=True)
```

```python
plt.figure(figsize=(4,3))
for col in df.columns:
  sns.distplot(df[col])
```

```python
with sns.axes_style('white'):
  sns.jointplot('x','y',df,kind='kde')
```

```python
iris=sns.load_dataset('iris')
sns.pairplot(iris,hue='species')
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',style='species')
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',kind='line',dashes=False,markers=True,ci=None,style='species')
```
