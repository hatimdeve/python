```python
import pandas  as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
import io
uploaded=files.upload()
```

```python
df=pd.read_csv(io.BytesIO(uploaded['test.csv']))
df.head()
```

```python
df['zscore']=(df.height-df.height.mean())/df.height.std()
df.head()
```

```python
df [df.zscore>3]
```

```python
df [df.zscore<-3]
```

```python
df[(df.zscore<-3) | (df.zscore>3)]
```

```python
df[(df.zscore>-3) & (df.zscore<3)]
```
