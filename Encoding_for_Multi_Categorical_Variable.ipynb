```python
import pandas as pd
df=pd.read_csv(io.BytesIO(uploaded['train.csv']),usecols=['X0','X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top10=[x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index ]
top10
```

```python
for label in top10:
   df[label]=np.where(df['X2']==label,1,0) 
df[['X2']+top10].head(40)
```

```python
def one_hot_topx(df,variable,top_x_labels):
  for label in top_x_labels:
   df[variable+'_'+label]=np.where(df[variable]==label,1,0) 
df=pd.read_csv(io.BytesIO(uploaded['train.csv']),usecols=['X0','X1','X2','X3','X4','X5','X6'])
one_hot_topx(df,'X2',top10)
df.head()
```
