```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'A':np.random.normal(0,1,1000),'B':np.random.normal(-2,2,1000),'C':np.random.normal(5,4,1000)})
```

```python
df.plot()
```

```python
df.plot.hist(alpha=0.5)
```

```python
df.plot.hist(stacked=True,bins=30,alpha=0.5)
```

```python
df2=pd.DataFrame(np.random.rand(10,4),columns=list('abcd'))
df2.plot.area()
```

```python
df3=pd.DataFrame(3*np.random.rand(4,2),index=list('abcd'),columns=list('xy'))
df3.plot.pie(subplots=True,figsize=(8,4))
```
