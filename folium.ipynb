```python
#conda install folium
import pandas as pd
import folium
from google.colab import files
import io
uploaded =files.upload()
```

```python
df=pd.read_csv(io.BytesIO(uploaded['global_covid19_mortality_rates.csv']))
df.head()
```

```python
m=folium.Map()
m
```

```python
m=folium.Map()
for i in range(0,len(df)):
  folium.Circle(location=[df.iloc[i]['Latitude'],df.iloc[i]['Longitude']],
                color='crimson',fill='crimson',radius=int(df.iloc[i]['Confirmed'])).add_to(m)
m
```
